# Car Price Exploratory Data Analysis

## Pakacges

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
car_prices_df = pd.read_csv ('Automobile_data.csv')
```

## View Data Set 

```python
# Dimension of data frame
car_prices_df.shape
# Create a copy as a standard practice
data = car_prices_df.copy()
print(f" There are {data.shape[0]} rows and {data.shape[1]} columns.")
data.head()
```

```python
# Correct the columns names & Data str format
## Uppercase & replace('-', '_')
data.columns = data.columns.str.capitalize()
data.columns = [str.replace('-',' ') for str in data.columns]
data.head()
```

```python
# Format the str in data
data = data.applymap(lambda s: s.capitalize() if type(s) == str else s)
data['Make'] = data['Make'].str.upper()
data['Aspiration'] = data['Aspiration'].str.upper()
data['Drive wheels'] = data['Drive wheels'].str.upper()
data['Fuel system'] = data['Fuel system'].str.upper()
data['Engine type'] = data['Engine type'].str.upper()
data.head()
```

```python
data.info() 
```

**Observations**

- There are 26 columns and 205 observations in the data set.
- Prelimnary assessment of data set shows "Normalized losses" has most missing values 
- There are several numeric data types stored as object (need to convert)
- Missing values to investigate further

## View Missing Values

```python
# Replace the "?" into NaN value in data set to find total null value by columns
data = data.replace('?', np.nan)
data.head()
```

```python
# To find total null value by columns
print('Total count of missing value by columns')
print()
for column in data.columns:
    if data[column].isnull().any():
        print (f"{column}", data[column].isnull().sum())     
```

**Observations**
- The above variables have missing valus along with total count
- The Normalized lossess has the highest missing values and all null to be dropped

```python
# Drop all null value
data = data.dropna(subset = ['Normalized losses','Num of doors','Bore',
                      'Stroke','Horsepower','Peak rpm','Price'])
data
```

## Data Format Treatment

```python
# Convert the Object into float and int respectively
data[['Bore','Stroke','Horsepower','Peak rpm','Price']] = data[['Bore','Stroke','Horsepower','Peak rpm','Price']].astype(float)
data[['Horsepower','Price','Peak rpm']] = data[['Horsepower','Price','Peak rpm']].astype('Int32')
data.info()
```

## To View Duplicate Values

```python
# Count duplicate value by column
data[data.duplicated()].count()
```

**Observaiton**
- There are no duplicate value

```python
data.describe().T
```

## Visualisations 

```python
ax, fig =plt.subplots(figsize=[15,7])
ax = sns.histplot(
    data,
    x='Make',
    hue='Body style',
    multiple='stack',
    # Add white borders to the bars.
    edgecolor='white',
    # Shrink the bars a bit so they don't touch.
    shrink=0.9,)

ax.set_title('Cars by Make and Body style')
plt.xticks(rotation=50)
# Remove 'Count' ylabel.
ax.set_ylabel(None);
```

```python
# List of number of cars by make 
data['Make'].value_counts()
```

**Key Insight**
- Top three volumns are coming from Toyota, Nissan and Honda respectively and has fair share of body style. 
- In Europian cars, only Mercedes-Benz has the mixed boady style. 

```python
ax, fig =plt.subplots(figsize=[15,7])
def plot_bar(data, col, col_y = 'Price'):
    sns.set_style("whitegrid")
    sns.barplot(x=col, y=col_y, ci=None, data=data)
    plt.xlabel(col) # Set text for the x axis
    plt.ylabel(col_y)# Set text for y axis
    plt.xticks(rotation=50)
    plt.show()
plot_bar(data, 'Make')  
```

**Key Insight**
- Toyota not only the top in volume yet higher priced car among Japanese cars.
- In European cars, Mercedes-Benz leading the price while Audi and BMW catching up each other.

```python
ax, fig =plt.subplots(figsize=[15,7])
def plot_box(data, col, col_y = 'Price'):
#     sns.set_style("whitegrid")
    sns.boxplot(x=col, y=col_y, data=data)
    plt.xlabel(col) # Set text for the x axis
    plt.ylabel(col_y)# Set text for y axis
    plt.show()
plot_box(data, 'Body style')  
```

```python
data['Count'] =1 # Setting up count 
group_data = data.groupby(['Body style','Make'])['Count'].sum() #Sum function
print(group_data)
print(data['Body style'].value_counts())
```

**Key Insight**
- Convertable is double the price as the mean value is twice compare to sedan but worth noticign the lowestm     volume.
- Toyota maintained it position in across all the boady style, let's look at Toyota price distribution by Body   style. 

```python
# Toyota data set copy
data_toyota = data[data.Make=='TOYOTA'].copy() # A copy data set only for Toyota
data_toyota['Price']=data_toyota['Price'].astype('float64') # Convert dtype of Price to float to plot violin

ax, fig =plt.subplots(figsize=[15,7])
sns.violinplot(x='Body style', y='Price', data=data_toyota);
```

**Key Insight (For Toyota)**
- Hatchback has the greater price distribution, stretched from low to premium price. It can be assumed that between ~13K and 15K price point is bottleneck, and it open above 15K



```python
data['Price']=data['Price'].astype('float64')
ax, fig =plt.subplots(figsize=[15,7])
ax = sns.violinplot(x='Num of cylinders', y='Price', hue='Fuel type',data=data, split=True)
```

**Key Insight**
- The low number of cylinder are mostly prefered due to low price regardless of fuel type.
- For five cylinders cars, diesel fuel types are expensive compare to gas

```python
def plot_scatter(data, cols, col_y = 'Price'):
    for col in cols:
        fig = plt.figure(figsize=(11,6)) # define plot area
        ax = fig.gca() # define axis   
        data.plot.scatter(x = col, y = col_y, ax = ax)
        ax.set_title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        ax.set_xlabel(col) # x axis label
        ax.set_ylabel(col_y)# y axis label
        plt.show()
num_cols = ['City mpg','Highway mpg']
plot_scatter(data, num_cols)      
```

**Key Insight**
- Both city and highway mpg, the lower the price the higher the mpg.
